**Inital Exploration and Cleaning**
We see some potentially erroneous data the following columns that warrant further investigation:

The num_of_pictures column seems to only contain the value 0, so we can drop this column.
Both the seller and offer_type columns contain the same value in almost all of the rows. These can be dropped as well. 



```python
#Removing $ and , from price column
autos["price"] = autos["price"].str.replace("$" , "").str.replace(",", "").astype(float)
print(autos["price"].head(7))

```

```python
#Removing km and , from the odometer column
autos["odometer"] = autos["odometer"].str.replace(",", "").str.replace("km", "").replace(" ", "")
autos["odometer"] = autos["odometer"].str.rstrip('km').astype(int)
                                                        


autos["odometer"] = autos["odometer"]
print(autos["odometer"].head())
```

```python
autos = autos.rename(columns = {"odometer" : "odometer_km"})
```

```python
#Exploring the price column for possible outliers

print("unique values:", autos['price'].unique().shape[0]) #shows unique values
print( autos['price'].describe()) #to view min/mex/median/mean
print(autos['price'].value_counts()) #Each price value and their counts


# In order to find the amount of highest price and lowest price and their counts,
#sorting can be done using series.sort_index() method.
print(autos['price'].value_counts().sort_index(ascending = True).head(20))
print(autos['price'].value_counts().sort_index(ascending = False).head(20))
```
